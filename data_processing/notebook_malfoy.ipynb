```python
def split_on_comma_outside_quotes(text):
    new_text = ""
    inside_quotes = False
    
    for char in text:
        if char == "'":
            inside_quotes = not inside_quotes
        if char == ',' and not inside_quotes:
            new_text.append('')
        if char == "[":
            new_text = new_text
        if char == "]":
            new_text = new_text
        else:
            new_text.append(char)
    
    return ''.join(new_text)
```

```python
ingredients_to_search = ['basmati rice', 'water', 'salt', 'chicken']

pattern = ''.join([f'(?=.*\\b{ingredient}\\b)' for ingredient in ingredients_to_search])

matching_rows = RAW_recipes[RAW_recipes['ingredients'].str.contains(pattern, case=False, na=False, regex=True)]

```

```python
matching_rows["full_nutrition"] = matching_rows['nutrition'].apply(ast.literal_eval)
matching_rows["full_nutrition"] = matching_rows['full_nutrition'].apply(lambda x: x)
matching_rows["calories"] = matching_rows['full_nutrition'].apply(lambda x: x[0])

max_calories = 500

filtered_recipes = matching_rows[matching_rows['calories'] < max_calories]

display(filtered_recipes.head())
```
