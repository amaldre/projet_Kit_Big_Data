```python
import pandas as pd
```

```python
df = pd.read_csv("../data/cloud_df.csv")
```

```python
df = df.sort_values(by="calories", ascending=False)
```

```python
df = df[df["calories"]<20000]
```

```python
df.head()
```

```python
df.sort_values(by="minutes", ascending=False)
```

```python
df = df[df["minutes"]<20000]
print(df.columns)
df = df.drop(columns=["contributor_id","user_id"])
```

```python
df = df[["name","mean_rating","comment_count","submitted","minutes","ingredients_replaced","calories","techniques","n_steps","date","recipe_id"]].rename(columns={'mean_rating':'Note moyenne', 'comment_count':'Nombre de commentaires', 'name':'Nom',
       'submitted':'Date de publication de la recette', 'n_steps':"Nombre d'étapes", 'date':"Dates des commentaires", 'ingredients_replaced':"Ingrédients", "minutes":"Durée de la recette (minutes)", "calories":"Calories", "techniques":"Techniques utilisées"
       })
```

```python
df.to_csv("../data/clean_cloud_df.csv")
```
